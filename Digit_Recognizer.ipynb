Importing the libraries

```python
# Import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential,model_from_json
from tensorflow.keras.layers import Dense,Dropout
```

Importing the dataset

```python
# Loading the dataset
df1=pd.read_csv('train.csv')     # Training data
```

```python
# Number of rows and columns in both datasets
df1.shape
```

Data Preprocessing of training data

```python
# First five rows in training data
df1.head()
```

```python
# Checking null values in training data
df1.isnull().sum()
```

```python
# Convert float columns to int columns
for i in df1.columns:
   if df1[i].dtypes=='float64':
       df1[i]=df1[i].astype('int64')
df1.head()
```

Splitting into dependent and independent variables

```python
# Pixel columns will be independent columns and label column will be target column
x_train=df1.iloc[:,1:]
y_train=df1['label']
```

```python
# First five rows of independent columns
x_train.head()
```

```python
# First five rows of target column
y_train.head()
```

As it is a Multi-class Classification problem, One-hot encoding will be applied on the target column to get dummy variables.

```python
# One hot encoding of target column
y_train=pd.get_dummies(y_train)
y_train.head()
```

Building the ANN Model

```python
# Model building
model=Sequential()

# First hidden layer with number of input dimensions=784 and number of nodes=256
model.add(Dense(units = 256, kernel_initializer = 'uniform', activation = 'relu',input_dim=784))
# Drop-out 20 percent nodes in a layer
model.add(Dropout(0.2))

# Second hidden layer with number of nodes=256
model.add(Dense(units= 256,kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(0.2))

# Third hidden layer with number of nodes=256
model.add(Dense(units= 256,kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(0.2))

# Output layer with number of nodes=10
model.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'softmax'))
```

```python
# Compiling the ANN model by applying the stochastic gradient descent
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the model to the training data, in which 10 per-cent is validation data with number of epochs=100
model.fit(x_train,y_train,epochs=100,validation_split=0.1)
```

Loading test data

```python
test=pd.read_csv('test.csv')
test.head()
```

Data Preprocessing of test data

```python
test.isnull().sum()
```

```python
# Convert float columns to int columns
for i in test.columns:
   if test[i].dtypes=='float64':
       test[i]=df1[i].astype('int64')
test.head()
```

Prediction for test data

```python
ypred=model.predict(test)
ypred_final=ypred.argmax(axis=1)
ypred_final
```
